# Example 9.27 - Sam's Decision To Party Or To Relax

We suppose that Sam wants to make an informed decision about whether to party or relax over the weekend. 

Sam prefers to party, but is worried about getting sick. 

Such a problem can be modeled as an MDP with the following:

Two States:
- Healthy
- Sick
  
Two Actions:
- Relax
- Party

The whole problem set up can be found here: https://artint.info/2e/html2e/ArtInt2e.Ch9.S5.html#Ch9.Thmciexamplered27.

```python
from value_iteration_code import value_iteration
```

We will assume, in line with the inital set up, the following transition probabilities and rewards:

### Transition Probabilities

| State   | Action | P(healthy) | P(sick) |
|---------|--------|------------|---------|
| healthy | relax  | 0.95       | 0.05    |
| healthy | party  | 0.70       | 0.30    |
| sick    | relax  | 0.50       | 0.50    |
| sick    | party  | 0.10       | 0.90    |

### Rewards

| State   | Action | Reward |
|---------|--------|--------|
| healthy | relax  | 7      |
| healthy | party  | 10     |
| sick    | relax  | 0      |
| sick    | party  | 2      |

```python
# Example 9.27 data

S = ["healthy", "sick"]

A = ["relax", "party"]

P = {
    "healthy": {
        "relax": [("healthy", 0.95), ("sick", 0.05)],
        "party": [("healthy", 0.7), ("sick", 0.3)]
    },
    "sick": {
        "relax": [("healthy", 0.5), ("sick", 0.5)],
        "party": [("healthy", 0.1), ("sick", 0.9)]
    }
}

R = {
    ("healthy", "relax", "healthy"): 7,
    ("healthy", "relax", "sick"): 7,

    ("healthy", "party", "healthy"): 10,
    ("healthy", "party", "sick"): 10,

    ("sick", "relax", "healthy"): 0,
    ("sick", "relax", "sick"): 0,

    ("sick", "party", "healthy"): 2,
    ("sick", "party", "sick"): 2
}

pi, V, delta_list = value_iteration(S, A, P, R, gamma=0.9, epsilon=0.001)

print("policy =", pi)
print("values =", V)
print("iterations =", len(delta_list))
print("last delta =", delta_list[-1])
```

# Grid World Problem

We suppose that an agent is placed in a small grid world and must choose actions in order to move between states and maximise reward.

The agent’s movement is uncertain, so taking an action does not always lead to the intended next state.

Such a problem can be modeled as an MDP with the following:

Four States:
- TL
- TR
- BL
- BR

Available Actions:
- From TL: R, D
- From TR: L, D
- From BL: R, U
- From BR: none (terminal state)

State BR is terminal.

The transition probabilities and rewards are defined by the table for the Grid World Problem below.

We will assume, in line with the initial set up, the following transition probabilities and rewards:

### Transition Probabilities

| State | Action | P(TL) | P(TR) | P(BL) | P(BR) |
|------|--------|-------|-------|-------|-------|
| TL   | R      | 0     | 0.9   | 0.1   | 0     |
| TL   | D      | 0     | 0.1   | 0.9   | 0     |
| TR   | L      | 0.9   | 0     | 0     | 0.1   |
| TR   | D      | 0.2   | 0     | 0     | 0.8   |
| BL   | R      | 0.1   | 0     | 0     | 0.9   |
| BL   | U      | 0.8   | 0     | 0     | 0.2   |

### Rewards

| State | Action | s' | Reward |
|------|--------|----|--------|
| TL   | R      | TR | -1     |
| TL   | R      | BL | -2     |
| TL   | D      | BL | -2     |
| TL   | D      | TR | -1     |
| TR   | L      | TL | -3/2   |
| TR   | L      | BR | 10     |
| TR   | D      | BR | 15     |
| TR   | D      | TL | -1     |
| BL   | R      | BR | 20     |
| BL   | R      | TL | -5/2   |
| BL   | U      | TL | -1/2   |
| BL   | U      | BR | 5      |
    

To solve this MDP, we use an implementation of the value iteration algorithm defined in a separate Python file.

The function `value_iteration_algorithm` computes the optimal value function and corresponding policy given:

- a set of states \( S \)
- a set of actions \( A \)
- a transition function \( P \)
- a reward function \( R \)
- a discount factor \( \gamma \)
- a convergence threshold \( \epsilon \)

We import this function as follows:

```python
from value_iteration_code import value_iteration
```

We then set the problem us as follows:

```python
# Grid world data

S = ["TL", "TR", "BL", "BR"]

A = ["R", "D", "L", "U"]

P = {
    "TL": {
        "R": [("TR", 0.9), ("BL", 0.1)],
        "D": [("BL", 0.9), ("TR", 0.1)]
    },
    "TR": {
        "L": [("TL", 0.9), ("BR", 0.1)],
        "D": [("BR", 0.8), ("TL", 0.2)]
    },
    "BL": {
        "R": [("BR", 0.9), ("TL", 0.1)],
        "U": [("TL", 0.8), ("BR", 0.2)]
    }
}

R = {
    ("TL", "R", "TR"): -1,
    ("TL", "R", "BL"): -2,

    ("TL", "D", "BL"): -2,
    ("TL", "D", "TR"): -1,

    ("TR", "L", "TL"): -3/2,
    ("TR", "L", "BR"): 10,

    ("TR", "D", "BR"): 15,
    ("TR", "D", "TL"): -1,

    ("BL", "R", "BR"): 20,
    ("BL", "R", "TL"): -5/2,

    ("BL", "U", "TL"): -1/2,
    ("BL", "U", "BR"): 5
}

pi, V, delta_list = value_iteration(S, A, P, R, gamma=0.9, epsilon=1e-6)

print("policy =", pi)
print("values =", V)
print("iterations =", len(delta_list))
print("last delta =", delta_list[-1])
```
